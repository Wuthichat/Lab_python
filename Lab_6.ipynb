```python
#Lab 6-1
import numpy as np
import matplotlib.pyplot as plt


Nbit = 1800
Nsamp = 10  #Eb
SNR = 10+0.1*8

np.random.seed(30)
a = np.random.randint(0,2, size=Nbit)
b = 2*a -1

plt.figure(1)
plt.stem(a)


plt.figure(2)
plt.stem(b)


x_t = []
for i in range(Nbit):
    x_t.extend([((-1)**a[i]) * -1] * Nsamp)

plt.figure(3)
plt.plot(x_t)
plt.show()
```

```python
#Lab 6-2

mu = 0
sigma  = np.sqrt(0.5*Nsamp*10**(-SNR/10))
n_t = np.random.normal(mu, sigma, len(x_t))

print("SNR(dB) : ",SNR)
print("sigma : ",sigma)
plt.figure(4)
plt.plot(n_t)
plt.show()
```

```python
#Lab 6-3
r_t = x_t + n_t
plt.figure(5)
plt.plot(r_t)
plt.show()
```

```python
#Lab 6-4
s_NRZL = ([1]*10)

z = []
for i in range(Nbit):
  z_t = np.multiply(r_t[i*Nsamp:(i+1)*Nsamp],s_NRZL)
  z_t_out = sum(z_t)
  z.append(z_t_out)

plt.figure(6)
plt.plot(z)
plt.figure(7)
plt.stem(z)

plt.figure(8)
plt.hist(z,bins = 100)
plt.show()
```

```python
#Lab 6-5

a_hat = [] #a^
for i in z:
  if i > 0:
    a_hat.append(1)
  else:
    a_hat.append(0)

err_num = sum((a!=a_hat))
print("error number : ",err_num)
print("BER : ",err_num/len(a))
print("SRN(dB) : ",SNR)
```
