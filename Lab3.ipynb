```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
N = 25000
n = np.random.normal(0, 2, size=N)
#n. np.random.normal(u,e,NumberSimple);u = center, e = standard deviation
a = 8
x = 2.5
b = 0


print("a:", a)
print("b:", b)

# Mean and variance of n
MeanValue = np.mean(n)
VarValue = np.std(n)

print("MeanValue:", MeanValue)
print("VarValue:", VarValue)

def output(n_normal):
    plt.figure(1)
    plt.plot(n_normal)
    plt.ylabel('n_Normal')
    plt.xlabel('k')

    plt.figure(2)
    plt.hist(n_normal, density=True, bins=40)
    plt.ylabel('Probability')
    plt.xlabel('n_normal')

    plt.show()

output(n)

def compute_prob(x, a, b):
    N = len(x)
    Z =[]
    for i in range(N):
        if x[i] > a and x[i] < b:
            Z.append(x[i])
        else:
            continue
    lenn = len(Z)
    print("lenn:", lenn)
    Prob = lenn / N
    print("Prob:", Prob)

    return Prob

# Pass the array x to compute_prob
print("Probability:", compute_prob(n, -2, 2))
```
