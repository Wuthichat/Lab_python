```python
import numpy as np
import matplotlib.pyplot as plt

A = 8
np.random.seed(A)
a = np.random.randint(0, 2, size=15)
print(a)

N = len(a)
x = 2 * a - 1
R = np.correlate(x, x, mode='full')
R_norm = R / np.max(R)

plt.stem(np.arange(-N+1, N), R_norm)
plt.figure(1).set_figwidth(12)
plt.title("Autocorrelation ID = 660XXXXXXXX , XXXXXXXXXXX ")
plt.xlabel("Time Lag")
plt.ylabel("RX(time lag)")
plt.grid(True, which='both')
plt.show()
```
