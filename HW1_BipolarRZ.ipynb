```python
''' C) plot spectrum to see both positive and negative side '''
import numpy as np
import matplotlib.pyplot as plt
import scipy.fftpack

Nbits = 10         # No. bits
Nsamp = 8 + round(8/2)  # No. samples/bits
pulse = 1

# 1. Generate data bits a, b #
np.random.seed(20 + 8)
a = np.random.randint(0, 2, Nbits)
b = 2 * a - 1

# 2. Generate modulated signal m(t) #
m = []
for i in range(Nbits):
    if a[i] == 1:
        m.extend([pulse] * (Nsamp // 2))
        m.extend([0] * (Nsamp // 2))
        pulse = pulse * -1
    else:
        m.extend([0] * Nsamp)

print("Data bits (a):", a)
print("Modulated signal (m):", m)

# Compute FFT of the modulated signal
Mf = scipy.fftpack.fft(m)
AbsMf = np.abs(Mf)
N = len(m)

# Generate frequency vector including both positive and negative frequencies
f = np.fft.fftfreq(N, d=1.0 / Nsamp)

# Shift the zero-frequency component to the center
Mf_shifted = np.fft.fftshift(Mf)
AbsMf_shifted = np.abs(Mf_shifted)
f_shifted = np.fft.fftshift(f)

# Plot the spectrum
plt.figure()
plt.plot(f_shifted, AbsMf_shifted)
plt.title('Spectrum of Modulated Signal - Wuthichat Phoopimon ID. 66015188')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()

```

```python
''' D) plot spectrum to see only positive side '''
import numpy as np
import matplotlib.pyplot as plt
import scipy.fftpack

Nbits = 10         # No. bits
Nsamp = 8 + round(8/2)  # No. samples/bits
pulse = 1

# 1. Generate data bits a, b #
np.random.seed(20 + 8)
a = np.random.randint(0, 2, Nbits)
b = 2 * a - 1

# 2. Generate modulated signal m(t) #
m = []
for i in range(Nbits):
    if a[i] == 1:
        m.extend([pulse] * (Nsamp // 2))
        m.extend([0] * (Nsamp // 2))
        pulse = pulse * -1
    else:
        m.extend([0] * Nsamp)

print("Data bits (a):", a)
print("Modulated signal (m):", m)

# Compute FFT of the modulated signal
Mf = scipy.fftpack.fft(m)
AbsMf = np.abs(Mf)
N = len(m)

# Generate frequency vector for positive frequencies
f = np.fft.fftfreq(N, d=1.0 / Nsamp)
positive_freq_indices = np.where(f >= 0)
f_positive = f[positive_freq_indices]
AbsMf_positive = AbsMf[positive_freq_indices]

# Plot the spectrum for positive frequencies
plt.figure()
plt.plot(f_positive, AbsMf_positive)
plt.title('Positive Spectrum of Modulated Signal - Wuthichat Phoopimon ID. 66015188')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```
